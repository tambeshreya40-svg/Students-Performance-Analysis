# 1. Gender Distribution
This graph shows the distribution of students by gender.

```python

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# 1. Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
print("-"*252)
```

# 2. Average Scores by Gender
This graph shows the Average Scores by gender.

```python
# 2. Average scores by gender
avg_scores = df.groupby('gender')[['math score','reading score','writing score']].mean().reset_index()
avg_scores.plot(x='gender', kind='bar', figsize=(8,5), title="Average Scores by Gender")
plt.ylabel("Score")
plt.show()
```

```python
print("-"*252)
```

# 3. Test Preparation course Effect 
This graph shows the Maths Scores by Test Preparation Course.

```python
# 3. Test preparation course effect
plt.figure(figsize=(6,4))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title("Math Score by Test Preparation Course")
plt.show()
```

```python
print("-"*252)
```

```python
# 4. Correlation Heatmap
This graph shows the Correlation between Scores.
```

```python
# 4. Correlation heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Scores")
plt.show()
```

```python
print("-"*252)
```

# 5. Histogram of Maths Score
This Graph shows the Maths score Distribution

```python
# Histogram of Math Score
plt.figure(figsize=(6,4))
sns.histplot(df['math score'], bins=20, kde=True)
plt.title("Math Score Distribution")
plt.show()
```

```python
print("-"*252)
```

## Conclusion
- Math, reading and writing scores are strongly correlated.
- Students who completed test preparation scored higher.
- Female students performed slightly better in language-related subjects.
- Parental education has a positive impact on scores.
